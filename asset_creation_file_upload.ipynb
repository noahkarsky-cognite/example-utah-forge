```python
UTAH_FORGE_DATASET_ID  = 8185991842002062
UTAH_FORGE_FILE_DATASET_ID = 2443662041002176
```

typically, a root asset will need to be created manually as it is more of a grouping concept one can use to organize assets.

```python
#make a root asset for all the wells
root = c.assets.retrieve(external_id='Utah Forge Wells')
if root is None:
    root_asset = Asset(name="Utah Forge Wells", external_id='Utah Forge Wells',    data_set_id=UTAH_FORGE_DATASET_ID)
    root_asset = c.assets.create(root_asset)
    print("Created root asset")
else:
    print("Root asset already exists")
```

Now that we have the root asset we can add our wells under it. As with many things with Python, there are many ways to go about this. Since we only have three assets to create I am going to just do it manually using dictionaries. I also had no structured way of getting this data, so I simply copied it from the end of well report and various supporting documentation. 

```python
#creating our first well asset under the root asset. It is the 68-32
well_data = {
    "Area": "Milford, Utah",
    "Field": "FORGE,Utah",
    "Operator": "University of Utah",
    "Well Name": '68-32',
    "Well Type": "Shallow Monitoring Well",
    "Lat": 38.501591,
    "Long": -112.886654,
    "CRS": "WGS84 UTM Zone 11",
    "ref_elevation": 5.7,
    "ref_elevation_unit": "ft",
    "TD":1000,
    "TD_unit":"ft",
}

#create a well asset
well_asset = Asset(name="68-32", data_set_id=UTAH_FORGE_DATASET_ID, parent_id=root_asset.id, metadata=well_data,external_id="68-32")
well_asset = c.assets.create(well_asset)

```

```python
folder_path = r"1153_drilling_data_well_loc/2019-well-68-32"
# make a list of all files in all folders in folder_path
file_list = []
for root, dirs, files in os.walk(folder_path):
    for file in files:
        file_list.append(os.path.join(root, file))

# we could call file upload once for all files, but I wanted to use mimetypes to get the mime type
[
    c.files.upload(
        x,
        data_set_id=UTAH_FORGE_FILE_DATASET_ID,
        asset_ids=[2698410095292759],
        mime_type=mimetypes.guess_type(x)[0],
    )
    for x in file_list
]
```
